```python
import pandas as pd
import matplotlib
import sklearn
import requests
import io
from bs4 import BeautifulSoup
```

```python
url = 'https://cocl.us/Geospatial_data'
response = requests.get(url).text
```

```python
data = io.StringIO(response)
```

```python
df = pd.read_csv(data)
```

```python
df
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_html = requests.get(url).text
soup = BeautifulSoup(wiki_html)
table = soup.find("table", attrs={"class":"wikitable sortable"})

# The first tr contains the field names.
headings = [th.get_text().strip() for th in table.find("tr").find_all("th")]

datasets = []
for row in table.find_all("tr")[1:]:
    dataset = dict(zip(headings, (td.get_text().strip() for td in row.find_all("td"))))
    datasets.append(dataset)

df1 = pd.DataFrame(data = datasets, columns = ['Postal Code', 'Borough', 'Neighborhood'])
df1 = df1[df1['Borough'] != 'Not assigned']
```

```python
df.head()
```

```python
df1.head()
```

```python
df2 = pd.merge(df1, df, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False,
         suffixes=('_x', '_y'),indicator=False,
         validate=None)
```

### Analysis of the borough data for Toronto 

```python
import folium # map rendering library
```

```python
latitude = 43.753259
longitude = -79.329656
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for ind,row in df2.iterrows():
    neighborhood = row['Neighborhood']
    borough = row['Borough']
    lat = row['Latitude']
    lng = row['Longitude']
    
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
